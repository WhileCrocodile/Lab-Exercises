```python
import requests

url = "https://gutendex.com/books"

# Find and return books with authors from the 6th century
start_year = "?author_year_start=501"
end_year = "&author_year_end=600"
year_url = url + start_year + end_year
year_request = requests.get(year_url).json()
print(year_request)

# Finds and returns all books with authors or title mentioning both the words "green" and "blue"
search = "?search=green%20blue"
search_url = url + search
search_request = requests.get(search_url).json()
print(search_request)

# Finds and returns all books with the topic of "food"
topic = "?topic=food"
topic_url = url + topic
topic_request = requests.get(topic_url).json()
print(topic_request)

# Finds and returns a list of all books, sorted by popularity
sort = "?sort=popular"
sort_url = url + sort
sort_request = requests.get(sort_url).json()
print(sort_request)

# Finds and returns a list of books written in Chinese
language = "?languages=zh"
language_url = url + language
language_request = requests.get(language_url).json()
print(language_request)
```

* Create some sort of program that uses the API (3pts)

```python
import requests
import csv
import pandas

csv_file = open("gutenberg_books.csv", "w", newline="", encoding="utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["Title", "Author(s)"])

url = "https://gutendex.com/books"

# Find and return books with authors from the 6th century
start_year = "?author_year_start=501"
end_year = "&author_year_end=600"
year_url = url + start_year + end_year
year_request = requests.get(year_url).json()
print("Count:", year_request["count"], "books")
for book in year_request["results"]:
    title = book["title"]
    authors = ""
    for author in book["authors"]:
        authors += author["name"] + ", "
    csv_writer.writerow([title, authors])
csv_file.close()

sheet = pandas.read_csv("gutenberg_books.csv")
print(sheet.to_string())
```
