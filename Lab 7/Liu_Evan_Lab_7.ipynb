```python
import requests
import pandas
import csv
from matplotlib import pyplot as plt
import numpy as np

class WeatherAPI:
    def __init__(self):
        self.url = "https://api.weather.gov/points/{lat},{lon}"
    
    def get_weather_metadata(self, lat, lon):
        weather_url = self.url.format(lat=lat, lon=lon)
        return requests.get(weather_url).json()
    
    def get_weather_forecast(self, lat, lon):
        weather_json = self.get_weather_metadata(lat, lon)
        forecast_json = requests.get(weather_json["properties"]["forecast"]).json()
        return forecast_json
    
    def get_weather_forecast_periods(self, lat, lon):
        '''
        Returns a dictionary of all weather periods, from now up to 7 days into the future.
        '''
        forecast_json = self.get_weather_forecast(lat, lon)
        return forecast_json["properties"]["periods"]

def main():
    # Binghamton has coordinates 42.0894° N, 75.9695° W
    lat = 42.0894
    lon = -75.9695
    
    WeatherAPIProxy = WeatherAPI()
    my_forecast = WeatherAPIProxy.get_weather_forecast_periods(lat, lon)

    weather_csv = open("weather_data.csv", "w", newline="", encoding="utf-8")
    weather_csv_writer = csv.writer(weather_csv)
    weather_csv_writer.writerow(["Time", "Temperature", "Temperature Unit", "Forecast Description"])
    
    for period in my_forecast:
        name = period["name"]
        temp = period["temperature"]
        temp_unit = period["temperatureUnit"]
        detailed_forecast = period["detailedForecast"]
        weather_csv_writer.writerow([name, temp, temp_unit, detailed_forecast])
    weather_csv.close()
   
    weather_sheet = pandas.read_csv("weather_data.csv")
    weather_dataframe = pandas.DataFrame(weather_sheet)
    
    day = weather_dataframe["Time"]
    temperature = weather_dataframe["Temperature"]
    
    # Bonus!
    # Adjusts the length of the x-axis to prevent text crowding
    # Source: APIs Day Two slides
    plt.figure(figsize = (25, 10))
    
    plt.bar(day, temperature)
    plt.xlabel("Day")
    plt.ylabel("Temperature")
    plt.title("Temperature Forecast (F)")
    
    plt.show()
                            
main()
```

* Read through the documentation of this public holiday API. There are seven different calls listed there, use all seven of them. Use comments to explain how the API is working & what each call is accomplishing. (You’ll need to use country codes) (4pts). 

Note that we store the functions we use to call the API in a class named PublicHolidayAPI. We explain how the API is working through comments above these class functions/methods, and not when we actually call the functions.

```python
import requests

class PublicHolidayAPI:
    def __init__(self):
        self.url = "https://date.nager.at/api/v3"
    
    # Slightly shortens the request function syntax
    def request_json(self, url):
        return requests.get(url).json()
    
    # Returns the information for a single country of our choosing (as a dictionary)
    def get_country_info(self, country_code):
        request_url = self.url + f"/CountryInfo/{country_code}"
        return self.request_json(request_url)
    
    # Returns a list of all countries with their country code and name (as dictionaries)
    def get_all_countries(self):
        request_url = self.url + "/AvailableCountries"
        return self.request_json(request_url)
    
    # Returns a list of all long weekends for a given year, country (as dictionaries)
    # Note that "needBridgeDay" indicates if you need to personally book a day off in-between to make
    # it a long weekend
    def get_long_weekends(self, year, country_code):
        request_url = self.url + f"/LongWeekend/{year}/{country_code}"
        return self.request_json(request_url)

    # Returns a list of all public holidays for a given year, country (as dictionaries)
    def get_public_holidays(self, year, country_code):
        request_url = self.url + f"/PublicHolidays/{year}/{country_code}"
        return self.request_json(request_url)
    
    # Returns a Response Code indicating if today is currently a holiday in a given country
    # This functions interprets that and returns a True/False. 200 means holiday, 204 means not
    # Based on UTC time, hence may need to offset manually to be accurate
    def get_holiday_today(self, country_code, offset=0):
        request_url = self.url + f"/IsTodayPublicHoliday/{country_code}?offset={offset}"
        today_holiday = requests.get(request_url)
        if today_holiday.status_code == 200:
            return True
        elif today_holiday.status_code == 204:
            return False
        else:
            return None
    
    # Returns a list of all upcoming holidays for a given country (as dictionaries)
    def get_upcoming_holidays(self, country_code):
        request_url = self.url + f"/NextPublicHolidays/{country_code}"
        return self.request_json(request_url)
    
    # Returns a list of all upcoming holidays globally for the next 7 days (as dictionaries)
    def get_upcoming_holidays_week(self):
        request_url = self.url + "/NextPublicHolidaysWorldwide"
        return self.request_json(request_url)


country_code = "US"
year = "2023"

holiday_api = PublicHolidayAPI()
country_info = holiday_api.get_country_info(country_code)
all_countries = holiday_api.get_all_countries()
long_weekends = holiday_api.get_long_weekends(year, country_code)
public_holidays = holiday_api.get_public_holidays(year, country_code)
today_holiday = holiday_api.get_holiday_today(country_code)
upcoming_holidays = holiday_api.get_upcoming_holidays(country_code)
upcoming_holidays_week = holiday_api.get_upcoming_holidays_week()

results = {"USA Country Info":country_info, "All Countries":all_countries, \
           "USA Long Weekends":long_weekends, "USA 2023 Public Holidays":public_holidays, \
           "USA Today Holiday?":today_holiday, "USA Upcoming Holidays":upcoming_holidays, \
           "Global Upcoming Holidays":upcoming_holidays_week}

for category, json in results.items():
    print(category)
    print(json)
    print("\n")
```
