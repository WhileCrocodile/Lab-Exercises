First we write a function to determine the longest word in a list.

```python
#This function only returns the first longest word it encounters
def check_longest_word(list):
    longest_word = ""
    longest_length = 0 
    for i in range(len(list)):
        if len(list[i]) > longest_length:
            longest_length = len(list[i])
            longest_word = list[i]
    
    return longest_word


list = ["apple", "chestnut", "gargoyle", "pandas", "sheep", "raptor"]
longest_word = check_longest_word(list)
print(longest_word)
```

The function above returns "chestnut", as it is the first longest word in the list, but "gargoyle" is just as long.

Next, we write a class to allow students to create flashcards and review them.

```python
import json
import random

class Flashcards:
    def __init__(self, name, imported_cards={}):
        self.name = name
        if isinstance(imported_cards, dict):
            self.flashcards = imported_cards
        else:
            print("Invalid flashcard format. Please import a dictionary.")
            self.flashcards = {}
        
    def identify(self):
        return self.name
        
    def add_item(self, question, answer):
        self.flashcards.update({question: answer})
    
    def remove_item(self, question):
        self.flashcards.pop(question)
    
    def list_all(self):
        for item in self.flashcards.items():
            print(item)
    
    def list_front(self):
        for key in self.flashcards.keys():
            print(key)
    
    def list_back(self):
        for value in self.flashcards.values():
            print(value)

    def import_flashcards(self, newflashcards):
        if isinstance(newflashcards, dict):
            self.flashcards = newflashcards
        else:
            print("Invalid flashcard format. Please import a dictionary.")
    
    def review_flashcards(self, numcards=-1):
        # Creates a copy of the dictionary, so we can remove previously-asked questions
        review_flashcards = self.flashcards
        
        # If the player enters too many or negative cards, default to all cards
        # Otherwise, default is all cards
        if numcards < 0 or numcards > len(review_flashcards):
            numcards = len(review_flashcards)

        for i in range(numcards):
            random_question = random.choice(list(review_flashcards.keys()))
            user_answer = input(random_question).lower()
            real_answer = review_flashcards[random_question]
            if user_answer == real_answer.lower():      #Lowercase to make comparisons uniform
                print(f"Correct! The answer to {random_question} is {real_answer}!")
            else:
                print(f"Incorrect... The answer to {random_question} was {real_answer}.")
            review_flashcards.pop(random_question)


def main():
    # Just showing that the class methods work
    print("Creating flashcards...")
    sample_flashcards = Flashcards("Sample Flashcards")
    print(sample_flashcards.identify())
    print("\n")

    # Showing that we can add items
    print("Adding items...")
    sample_flashcards.add_item("Dog", "Animal")
    sample_flashcards.add_item("Pear", "Fruit")
    sample_flashcards.add_item("Cup", "Object")

    # Showing that we can list items
    print("Listing items...")
    sample_flashcards.list_all()
    print("\n")
    print("Listing front side...")
    sample_flashcards.list_front()
    print("\n")
    print("Listing back side...")
    sample_flashcards.list_back()
    print("\n")

    # Showing that we can remove items
    print("Removing items...")
    sample_flashcards.remove_item("Dog")
    sample_flashcards.list_all()
    print("\n")

    # Showing that we can import items
    print("Importing new flashcards...")
    sample_flashcards.import_flashcards({"Airplane": "Vehicle"})
    sample_flashcards.list_all()
    print("\n")

    # Now, into the Pokemon type review game
    with open("pokemon_primary_types.txt") as pokemon_dict:
        pokemon_dict = json.loads(pokemon_dict.read())
    pokemon_flashcards = Flashcards("pokemon_flashcards", pokemon_dict)
    while True:
        try:
            numcards = int(input("How many cards would you like to review today? (Enter -1 for all)"))
            break
        except:
            print("Invalid input! Try again...")
    print("What is the primary type of these Pokemon?")
    pokemon_flashcards.review_flashcards(numcards)

main()
```

This is my completed Hangman game. 

I took some inspiration from one of my peers in class who used turtle to draw her hangman. I instead chose to use a dictionary with six different ASCII art versions of my hangman in various states. Each time the player loses a guess/life, the appropriate hangman is pulled from the dictionary, and printed for the player.

I originally wrote my game as an extremely long function, but I decided it looks and works better as a class.

Instead of giving players the position where the characters were found, I decided a word display was enough. If they want to be precise, they could always just count manually.

I noticed in class, that there was an implementation where the game displays words by filling in the underscores where they match already-guessed words. Instead of doing that, I maintain two lists. One list has the real word, the other list begins as just underscores, and letters are slowly copied over from the real-word-list to the in-progress-word-list.

I also don't restrict players to the English alphabet. That is to say, "24%sa" is a valid, albeit always-incorrect, input.

```python
import random
import json

sowpods = "sowpods.txt"

def word_generator():
    filename = sowpods
    f = open(filename, "r", encoding="utf8")
    words = f.readlines()
    f.close()

    return random.choice(words).strip()

# Implements Hangman as a class.
class Hangman:
    
    def __init__(self):
        # Initializes the variables used in the game so they can be
        # easily accessed by class methods.
        self.game_running = False

        with open("hangman_states.json") as hangman_states:
            self.hangman_dict = json.loads(hangman_states.read())
        
        self.reset_game()

    def reset_game(self):
        # Reset the game variables at game restart/end
        self.actual_word = ""
        self.guess = ""
        self.word_length = 0
        self.number_guesses = 6

        # Creates lists to hold letters, for easy manipulation
        self.actual_word_list = []        
        self.known_word_list = []
        
        # Variable for reconverting list back into string, for printing.
        self.known_word_string = ""
        
        # Stores words already guessed by the player
        self.already_guessed = set()

        # Determines the state/outcome of the current game
        self.game_won = False
        self.game_over = False

    def ask_guess(self):
        # Asks for the player's guess
        self.guess = str(input("Guess a letter, or the whole word. ")).upper()

        # If they have previously made that self.guess, ask again. Otherwise, add to set of previous guesses.
        while self.guess in self.already_guessed:
            print("You already guessed that! Try another guess.\n")
            self.guess = str(input("Guess a letter, or the whole word. ")).upper()
        self.already_guessed.add(self.guess)

        return self.guess
    
    def evaluate_letter(self):
        # If player guesses correctly, update known_word_list using self.actual_word_list by 
        # finding and filling in the blanks. Then, update self.known_word_string.
        if self.guess in self.actual_word_list:
            letter_indices = [i for i in range(len(self.actual_word_list)) if self.actual_word_list[i] == self.guess]
            for i in letter_indices:
                self.known_word_list[i] = self.actual_word_list[i]
            self.known_word_string = "".join(self.known_word_list)
            print(f"Correct! Your word is {self.known_word_string} ({self.word_length} characters).\n")
        
        # Otherwise, if incorrect, do this. 
        else:
            self.number_guesses -= 1
            print(f"Incorrect! Your word is {self.known_word_string} ({self.word_length} characters).")
            print(f"You have {self.number_guesses} guesses remaining.\n")
            self.show_hangman()

    def evaluate_word(self):
        # If the player guesses correctly
            if self.guess == self.actual_word:
                print("Correct!\n")
                self.game_won = True
                self.game_over = True
            # If the player guesses incorrectly
            else:
                self.number_guesses -= 1
                print(f"Incorrect! Your word is {self.known_word_string} ({self.word_length} characters).")
                print(f"You have You have {self.number_guesses} guesses remaining.\n")
                self.show_hangman()
    
    def show_hangman(self):
        print(self.hangman_dict[str(self.number_guesses)])

    def start_game(self):
        self.game_running = True

        # self.game_running == True as long as the player does not refuse to play the game
        while self.game_running == True:
            self.run_game()

            restart = str(input("Would you like to play again? (y/n) ")).lower()
            
            while restart != "y" and restart != "n":
                restart = str(input("Invalid input. Would you like to play again? (y/n) ")).lower()
            if restart == "y":
                print("Sure! Restarting game...\n")    
            if restart == "n":
                print("Okay! Thanks for playing!\n")
                self.game_running = False
            
            self.reset_game()
    
    def run_game(self):
        # Chooses a random word and determines some things about it
        self.actual_word = word_generator()
        self.word_length = len(self.actual_word)

        print("FOR DEBUGGING PURPOSES: " + self.actual_word)
   
        # Populates both lists with appropriate elements
        for letter in self.actual_word:
            self.actual_word_list.append(letter)
            self.known_word_list.append("_")

        # Converts known_word_list into a string to be printed for the player
        self.known_word_string = "".join(self.known_word_list)
        
        # Let's play!
        print(f"Let's play Hangman! Your word is: {self.known_word_string} ({self.word_length} characters).")
        print(f"You have {self.number_guesses} guesses.")

        while self.game_over == False:
            # Asks the player for their self.guess (and stores their self.guess in "self.already_guessed")
            self.ask_guess()
            
            # If the player guesses a whole word (more than one character).
            if len(self.guess) != 1:
                self.evaluate_word()
            
            # If the player guesses a single letter
            else:
                self.evaluate_letter()
            
            # Flips the win flag if known_word_string matches the actual_word.
            # This should only trigger if the player won by filling in the blanks
            # (Not guessing the whole word at once)
            if self.known_word_string == self.actual_word:
                self.game_won = True
                self.game_over = True
            
            # Lose condition
            if self.number_guesses == 0:
                self.game_over = True
                print("You ran out of guesses! Game over.")
                print(f"Your word was {self.actual_word}.\n")
            
            # Win condition
            if self.game_won == True:
                print("You win!")
                print(f"Your word was {self.actual_word}. You had {self.number_guesses} guesses remaining.\n")

def main():
    hangman = Hangman()
    hangman.start_game()

main()
```
