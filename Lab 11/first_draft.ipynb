```python
import requests
import tkinter as tk
import random
import time

def get_trivia():
    num_questions = 10
    category = "film_and_tv"
    difficulty = "easy"
    url = f"https://the-trivia-api.com/api/questions?categories={category}&limit={num_questions}\
    &region=US&difficulty={difficulty}"
    return requests.get(url).json()

def advance_question():
    global trivia_questions
    global radio_buttons
    global question_display
    global correct_answer
    global questions_answered
    
    current_trivia = next(trivia_questions)
    question = current_trivia["question"]
    display_text = f"{questions_answered+1}. {question}"
    question_display.config(text=display_text)
    
    current_correct_answer = current_trivia["correctAnswer"]
    correct_answer = current_correct_answer
    incorrect_answers = current_trivia["incorrectAnswers"]
    all_answers = incorrect_answers.copy()
    all_answers.append(current_correct_answer)
    random.shuffle(all_answers)
    for index, button in enumerate(radio_buttons):
        current_answer = all_answers[index]
        button.config(text=current_answer, value=current_answer)
    current_selection.set(all_answers[0])

def submit_answer():
    global current_selection
    global correct_answer
    global questions_correct
    global questions_answered
    global feedback_label
    global correct_counter
    
    if current_selection.get() == correct_answer:
        feedback_label.config(text="That last answer was correct!")
        questions_correct += 1
    else:
        feedback_label.config(text="That last answer was incorrect...")
    questions_answered += 1
    
    correct_counter.config(text=f"{questions_correct} questions correct so far.")
    if questions_answered == 10:
        for widget in root.winfo_children():
            widget.destroy()
        text = f"Game over! You got {questions_correct}/10 questions correct!" 
        game_over_label = tk.Label(root, text=text)
        game_over_label.pack()

    advance_question()

trivia_questions = iter(get_trivia())

root = tk.Tk()
root.geometry("800x200")
root.title("Trivia Questions Game")
questions_answered = 0
questions_correct = 0
correct_answer = ""

question_display = tk.Label(text="1. Trivia Question", wraplength=500, padx=10, pady=10, justify="left")
question_display.pack(anchor=tk.W)

radio_buttons = []
dummy_answers = ["Answer1", "Answer2", "Answer3", "Answer4"]
current_selection = tk.StringVar()
for answer in trivia_answers:
    new_button = tk.Radiobutton(root, text=answer, value=answer, var=current_selection, padx=20)
    new_button.pack(anchor = tk.W)
    radio_buttons.append(new_button)
advance_question()

submit_button = tk.Button(text="Submit answer", command=submit_answer, padx=30)
submit_button.pack(anchor=tk.W)

feedback_label = tk.Label(text="Please choose an answer.", padx=20)
correct_counter = tk.Label(text="0 questions correct so far.", padx=20)
feedback_label.pack(anchor=tk.W)
correct_counter.pack(anchor=tk.W)

root.mainloop()
```
