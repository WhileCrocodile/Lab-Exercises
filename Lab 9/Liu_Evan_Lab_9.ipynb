> Part 2: Try and create your own parser for an API. If you are using an API for your project, build one for it; if you are not, choose an API that makes sense to you or is related to your project in some way. Include at least five functions for 1pt each. 

```python
import requests

# Parser for data from the Art Institute of Chicago
url = "https://api.artic.edu/api/v1/artworks"

def get_artwork_byid(art_id):
    request_url = url + f"/{art_id}"
    return requests.get(request_url).json()

def get_artwork_byid_desc(art_id):
    art_json = get_artwork_byid(art_id)
    title = art_json["data"]["title"]
    desc = art_json["data"]["thumbnail"]["alt_text"]
    artist = art_json["data"].setdefault("artist_display", "Unknown").split("\n")[0]
    place_of_origin = art_json["data"].setdefault("place_of_origin", "Unknown")
    return {"title": title, "desc":desc, "artist":artist, "place_of_origin":place_of_origin}
    
def get_artwork_bysearch(keyword):
    request_url = url + f"/search?q={keyword}"
    return requests.get(request_url).json()

def get_first_artwork_bysearch(keyword):
    request_url = url + f"/search?q={keyword}&size=1"
    return requests.get(request_url).json()

def get_first_artwork_bysearch_desc(keyword):
    art_json = get_first_artwork_bysearch(keyword)
    title = art_json["data"][0]["title"]
    desc = art_json["data"][0]["thumbnail"]["alt_text"]
    artist = art_json["data"][0].setdefault("artist_display", "Unknown").split("\n")[0]
    place_of_origin = art_json["data"][0].setdefault("place_of_origin", "Unknown")
    return {"title": title, "desc":desc, "artist":artist, "place_of_origin":place_of_origin}

# Prints the entire json; very messy
# print(get_artwork_byid(129884))

# Prints some key identifying information
print(get_artwork_byid_desc(129884))

# Prints the entire search results; also very messy
# print(get_artwork_bysearch("cats"))

# Prints the first json for the search
print(get_first_artwork_bysearch("cats"))

# Prints a desc of the first json
print(get_first_artwork_bysearch_desc("cats"))
```
